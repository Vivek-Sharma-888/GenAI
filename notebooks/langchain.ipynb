```python
from langchain_openai import ChatOpenAI
from langchain_core.prompts import ChatPromptTemplate
from langchain_core.output_parsers import StrOutputParser

out = StrOutputParser()

def upperResult(result:str):
    return result.upper()

prompt = [{"role":"system","content":"you are {language} translator and convert any text into {language}"},
          {"role":"user","content":"{query}"}]

llm = ChatOpenAI(model="gpt-4o")
prompts = ChatPromptTemplate.from_messages(prompt)
# final_prompt = prompts.format_messages(language="hindi",query="I have done lot of mistakes")
# res = llm.invoke(final_prompt)
# res.content
chains = prompts|llm|out|upperResult
res = chains.invoke({"language":"tamilish","query":"I have done lot of mistakes"})
res
```
